```python
from sklearn.utils import shuffle
from sklearn.preprocessing import normalize
from sklearn.naive_bayes import MultinomialNB, CategoricalNB
import scipy.io as sc
import numpy as np
import pandas as pd
import matplotlib.pyplot
import tensorflow as tf
from tensorflow.keras import datasets, layers, models

data_dir = "./drive/MyDrive/CPE646/"

train_frames = sc.loadmat(data_dir + 'TrainingRxFrames.mat')
train_labels = sc.loadmat(data_dir + 'TrainingRxLabels.mat')

valid_frames = sc.loadmat(data_dir + 'ValidationRxFrames.mat')
valid_labels = sc.loadmat(data_dir + 'ValidationRxLabels.mat')

test_frames = sc.loadmat(data_dir + 'TestRxFrames.mat')
test_labels = sc.loadmat(data_dir + 'TestRxLabels.mat')

total_data = sc.loadmat(data_dir + 'totalData_struct.mat')
```

```python
import sklearn.preprocessing
## x1 feature = inphase signal
## x2 feature = quadrature signal
train_x = train_frames['rxTrainFrames']
nb_X = train_x[0]
nb_X = np.reshape(nb_X, (88000, 2048))
nb_X = normalize(nb_X, axis=0)
print(np.shape(nb_X))

train_x = np.reshape(train_x, (88000, 2, 1024, 1)).astype("float32")
train_y = train_labels['rxTrainY']
train_y = train_y[0]
train_y = train_y[0][6]
nb_y = np.reshape(train_y, (88000,))
train_y = train_y - 1
train_y = tf.keras.utils.to_categorical(train_y, num_classes=11, dtype='float32')
train_x, train_y = shuffle(train_x, train_y)

valid_x = valid_frames['rxValidFrames']
#valid_x = valid_x[0]
valid_x = np.reshape(valid_x, (11000, 2, 1024, 1)).astype("float32")
valid_y = valid_labels['rxTestY']
valid_y = valid_y[0]
valid_y = valid_y[0][6]
valid_y = valid_y - 1
valid_y = tf.keras.utils.to_categorical(valid_y, num_classes=11, dtype='float32')
valid_x, valid_y = shuffle(valid_x, valid_y)

test_x = test_frames['rxTestFrames']
#test_x = test_x[0]
test_x = np.reshape(test_x, (11000, 2, 1024, 1)).astype("float32")
test_y = test_labels['TestRxLabels'][0]
test_y = test_y[0][6]
test_y = test_y - 1
test_y = tf.keras.utils.to_categorical(test_y, num_classes=11, dtype='float32')

#test_y = np.reshape(1, 88000)

#train_x = train_x.melt(id_vars=train_x[2])
print("\nPrinting next trainX.\n")
print(train_x)
print("\nPrinting next trainY.\n")
print(train_y)
print("\nPrinting next validX.\n")
print(valid_x)
print("\nPrinting next validY.\n")
print(valid_y)
print("\nPrinting next testX.\n")
print(test_x)
print("\nPrinting next testY.\n")
print(test_y)
print(train_x.shape)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Conv2D(32, 1, activation='relu', padding='same', 
                                 input_shape=(2,1024,1)))
model.add(tf.keras.layers.MaxPooling2D(1,2))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Conv2D(64, 1, activation='relu', padding='same'))
model.add(tf.keras.layers.MaxPooling2D(1,2))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Conv2D(128, 1, activation='relu', padding='same'))
model.add(tf.keras.layers.MaxPooling2D(1,2))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.AveragePooling2D((1,2)))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dense(256, activation='relu'))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(11, activation='softmax'))
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
history = model.fit(train_x, train_y, batch_size=256, epochs=12, validation_data=(valid_x, valid_y))
```

```python
results = model.evaluate(test_x, test_y)
print(results)
```

```python
from sklearn.cluster import kmeans_plusplus
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample data
n_samples = 4000
n_components = 4

X, y_true = make_blobs(
    n_samples=n_samples, centers=n_components, cluster_std=0.60, random_state=0
)
X = X[:, ::-1]

# Calculate seeds from kmeans++
centers_init, indices = kmeans_plusplus(X, n_clusters=4, random_state=0)

# Plot init seeds along side sample data
plt.figure(1)
colors = ["#4EACC5", "#FF9C34", "#4E9A06", "m"]

for k, col in enumerate(colors):
    cluster_data = y_true == k
    plt.scatter(X[cluster_data, 0], X[cluster_data, 1], c=col, marker=".", s=10)

plt.scatter(centers_init[:, 0], centers_init[:, 1], c="b", s=50)
plt.title("K-Means++ Initialization")
plt.xticks([])
plt.yticks([])
plt.show()
```
