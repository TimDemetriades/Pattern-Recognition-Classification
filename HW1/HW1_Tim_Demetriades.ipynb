#### 1. Sketch the likelihood function for each class (ref. Figure 2.1). Calculate the likelihood value for each class if x=5.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
import seaborn as sns

mu1 = 0
sigma1 = 1

mu2 = 2
sigma2 = 2    # variance = 4 (std dev squared)

x = np.linspace(-5, 5, 100)    # vector of evenly spaced nums from -10 to 10
```

**Gaussian Distribution Formula**



```python
likelihood1 = (1 / (1 * np.sqrt(2 * np.pi))) * (np.exp(-0.5 * ((x - 0) / 1) ** 2))
likelihood2 = (1 / (2 * np.sqrt(2 * np.pi))) * (np.exp(-0.5 * ((x - 2) / 2) ** 2))
```

```python
plt.plot(x, likelihood1, label = 'Class 1')
plt.plot(x, likelihood2, label = 'Class 2')
plt.legend(loc = "upper left", fontsize = 14)
plt.grid()
plt.title('Likelihood Functions')
```

```python
likelihood_val1 = likelihood1[np.where(x == 5)][0]
likelihood_val2 = likelihood2[np.where(x == 5)][0]
print('Likelihood values for Class 1 and Class 2 at x = 5:')
print(f'Class 1 - {likelihood_val1:.8f}')
print(f'Class 2 - {likelihood_val2:.8f}')
```

#### 2. Sketch the evidence distribution. Calculate the evidence value if x = 5.

**Evidence Formula**



```python
prior1 = 3 / 5
prior2 = 2 / 5

evidence = (prior1 * likelihood1) + (prior2 * likelihood2)
```

```python
plt.plot(x, evidence)
plt.grid()
plt.title('Evidence Distribution')
```

```python
evidence_val1 = evidence[np.where(x == 5)][0]
print(f'Evidence value at x = 5: {evidence_val1:.4f}')
```

#### 3. Sketch the posterior probability function for each class (ref. Figure 2.2). Calculate the posterior probability value for each class if x = 5.

**Posterior Probability Formula**



```python
posterior1 = (likelihood1 * prior1) / evidence
posterior2 = (likelihood2 * prior2) / evidence
```

```python
plt.plot(x, posterior1, label = 'Class 1')
plt.plot(x, posterior2, label = 'Class 2')
plt.legend(loc = "upper left", fontsize = 14)
plt.grid()
plt.title('Posterior Probability Functions')
```

```python
posterior_val1 = posterior1[np.where(x == 5)][0]
posterior_val2 = posterior2[np.where(x == 5)][0]
print('Posterior Probability values for Class 1 and Class 2 at x = 5:')
print(f'Class 1 - {posterior_val1:.8f}')
print(f'Class 2 - {posterior_val2:.8f}')
```

#### 4. Sketch the likelihood ratio function (ref. Figure 2.3). Calculate the likelihood ratio value if x=5.

**Likelihood Ration Function**



```python
likelihood_ratio = likelihood1 / likelihood2
```

```python
plt.plot(x, likelihood_ratio)
plt.grid()
plt.title('Likelihood Ratio Function')
```

```python
likelihood_ratio_val = likelihood_ratio[np.where(x == 5)][0]
print(f'Likelihood Ratio value at x = 5: {likelihood_ratio_val:.8f}')
```

**Conditional Risk**



```python
loss11 = 0
loss21 = 2
loss12 = 4
loss22 = 0

risk1 = (loss11 * posterior1) + (loss12 * posterior2)
risk2 = (loss21 * posterior1) + (loss22 * posterior2)
```

```python
plt.plot(x, risk1, label = 'Class 1')
plt.plot(x, risk2, label = 'Class 2')
plt.legend(loc = "upper left", fontsize = 14)
plt.grid()
plt.title('Bayes Risk')
```

```python
risk_val1 = risk1[np.where(x == 5)][0]
risk_val2 = risk2[np.where(x == 5)][0]
print('Bayes Risk values for Class 1 and Class 2 at x = 5:')
print(f'Class 1 - {risk_val1:.8f}')
print(f'Class 2 - {risk_val2:.8f}')
```
